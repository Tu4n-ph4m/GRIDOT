```python
import pandas as pd
import scanpy as sc
```

```python
sc.settings.verbosity = 3  # verbosity: errors (0), warnings (1), info (2), hints (3)
sc.logging.print_header()
sc.settings.set_figure_params(dpi=80, facecolor="white")
```

```python
results_file = "./data/kidney_rna.h5ad"  # the file that will store the analysis results
```

```python
rna_adata = sc.read_h5ad(results_file)
# rna_adata.var.index = rna_adata.var['Gene']
rna_adata
```

```python
from collections import Counter
Counter(rna_adata.obs['patient'])
```

```python
rna_adata = rna_adata[rna_adata.obs['patient'].isin(['control1'])]
rna_adata = rna_adata[rna_adata.obs['celltype'].isin(['CNT_PC', 'DCT', 'ENDO'])]
```

```python
Counter(rna_adata.obs['celltype'])
```

```python
# rna_adata.obs['cell_id'] = rna_adata.obs['orig.ident']
# rna_adata.obs.index = rna_adata.obs['cell_id']
```

```python
rna_adata.var_names_make_unique()  # this is unnecessary if using `var_names='gene_ids'` in `sc.read_10x_mtx`
```

```python
sc.pl.highest_expr_genes(rna_adata, n_top=20)
```

```python
sc.pp.filter_cells(rna_adata, min_genes=200)
sc.pp.filter_genes(rna_adata, min_cells=3)
rna_adata
```

```python
# annotate the group of mitochondrial genes as "mt"
rna_adata.var["mt"] = rna_adata.var_names.str.startswith("MT-")
sc.pp.calculate_qc_metrics(
    rna_adata, qc_vars=["mt"], percent_top=None, log1p=False, inplace=True
)
```

```python
sc.pl.violin(
    rna_adata,
    ["n_genes_by_counts", "total_counts", "pct_counts_mt"],
    jitter=0.4,
    multi_panel=True,
)
```

```python
sc.pl.scatter(rna_adata, x="total_counts", y="pct_counts_mt")
sc.pl.scatter(rna_adata, x="total_counts", y="n_genes_by_counts")
```

```python
# rna_adata = rna_adata[rna_adata.obs.n_genes_by_counts < 2500, :]
rna_adata = rna_adata[rna_adata.obs.pct_counts_mt < 5, :].copy()
rna_adata
```

```python
rna_adata.write_h5ad("./data/rna_data_filtered_pca_1control_3cells_real.h5ad")
```

```python
sc.pp.normalize_total(rna_adata, target_sum=1e4)
```

```python
sc.pp.log1p(rna_adata)
```

```python
sc.pp.highly_variable_genes(rna_adata, min_mean=0.0125, max_mean=3, min_disp=0.5)
```

```python
# rna_adata.raw = rna_adata.copy()
```

```python
rna_adata = rna_adata[:, rna_adata.var.highly_variable]
```

```python
sc.pp.regress_out(rna_adata, ["total_counts", "pct_counts_mt"])
```

```python
sc.pp.scale(rna_adata, max_value=10)
```

```python
sc.tl.pca(rna_adata, svd_solver="arpack")
```

```python
sc.pl.pca(rna_adata, color="celltype")
```

```python
sc.pp.neighbors(rna_adata)
```

```python
sc.tl.umap(rna_adata)
```

```python
sc.pl.umap(
    rna_adata,
    color="celltype",
    # Setting a smaller point size to get prevent overlap
    size=10,
)
```
